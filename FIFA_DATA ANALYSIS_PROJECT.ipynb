```python
# Splitting the datasets into categorical and numerical dataframe
categorical_df = df[["Name", "Nationality",'Age', 'Overall','Preferred Positions']]
```

```python
numeric_df = df[['Acceleration','Aggression',
       'Agility', 'Balance', 'Ball control', 'Composure', 'Crossing', 'Curve',
       'Dribbling', 'Finishing', 'Free kick accuracy', 'GK diving',
       'GK handling', 'GK kicking', 'GK positioning', 'GK reflexes',
       'Heading accuracy', 'Interceptions', 'Jumping', 'Long passing',
       'Long shots', 'Marking', 'Penalties', 'Positioning', 'Reactions',
       'Short passing', 'Shot power', 'Sliding tackle', 'Sprint speed',
       'Stamina', 'Standing tackle', 'Strength', 'Vision', 'Volleys']]
```

```python
numeric_df["Acceleration"].unique()
```

```python
numeric_df = numeric_df.apply(lambda x: x.str.split('[+-]').str[0])
numeric_df = numeric_df.apply(lambda x: pd.to_numeric(x))

```

```python
numeric_df
```

```python
# Checking for outliers in our datasets
def replace_outliers(dataframe):
    for column_name in dataframe.columns:
        # Calculate quartiles and IQR
        Q1 = np.quantile(dataframe[column_name], 0.25)
        Q3 = np.quantile(dataframe[column_name], 0.75)
        IQR = Q3 - Q1

        # Calculate upper and lower limits
        lower_limit = Q1 - 1.5 * IQR
        upper_limit = Q3 + 1.5 * IQR

        # Replace outliers with NaN
        dataframe.loc[(dataframe[column_name] < lower_limit) | (dataframe[column_name] > upper_limit), column_name] = np.nan

    return dataframe


```

```python
# THERE IS SO MANY OUTLIERS IN THE NUMERICAL DATASET
numeric_df = replace_outliers(numeric_df)
```

```python
numeric_df.isna().sum()
```

```python
# fill all the nan with the median of each column

def replace_nan(dataframe):
    for column_name in dataframe.columns:
        # Calculate median
        median_val = dataframe[column_name].median()
        dataframe[column_name].fillna(median_val,inplace =True)
        

    return dataframe
```

```python
numeric_df = replace_nan(numeric_df)

numeric_df.isna().sum()
```

```python

final_df = pd.concat([categorical_df, numeric_df], axis=1)
```

```python
# Drop all duplicates in final_df

final_df.drop_duplicates(inplace=True)
```

```python
top_positions = final_df['Preferred Positions'].head(10)

top_positions_df = pd.DataFrame({'Preferred Positions': top_positions})
order = top_positions.value_counts().index


# Create a catplot using Seaborn
sns.catplot(x='Preferred Positions', kind='count', data=top_positions_df, height=6, aspect=1.5, order= order)

# Set plot labels and title
plt.xlabel('Preferred Positions')
plt.ylabel('Count')
plt.title('Count of Top 10 Preferred Positions')

plt.show()
```
